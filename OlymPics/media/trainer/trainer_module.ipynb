```python
from clustimage import Clustimage
from PIL import Image
import cv2

cl = Clustimage(method='pca')

face_results = cl.extract_faces('1.jpg')

for i in range(2,1267,1):
    face_results1 = cl.extract_faces(f'{i}.jpg')
    face_results['pathnames_face'].append(face_results1['pathnames_face'][0])
    face_results['filenames'].append(f'{i}.jpg')
```

```python
results = cl.fit_transform(face_results['pathnames_face'])
```

```python
cl.save('clustimage_model1', overwrite=True)
```
